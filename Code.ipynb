```python
# supress Warnings

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
```

```python
# reading the Transaction data

df = pd.read_excel('Transactional Data.xlsx',sheet_name = 'Transaction data (Problem 1)')
df
```

#### Split Transactions

```python
# creating a dataset of the split transactions

single = df[['From Account','To Account']].drop_duplicates(keep=False)
split = df[~df.index.isin(single.index)]
split
```

```python
# reading the transaction limits

limit = pd.read_excel('Transactional Data.xlsx',sheet_name = 'Transaction Limit')
limit.rename(columns = {'Transaction Type':'Type of Transaction'},inplace = True)
limit
```

```python
# changing the date time format

split['Transaction Date & Time'] = pd.to_datetime(split['Transaction Date & Time'])
split['Transaction Date & Time'] = [d.date() for d in split['Transaction Date & Time']]
split
```

```python
split['KO ID / Type of Transaction'] = split['KO ID'] + ' / ' + split['Type of Transaction']
split.drop(['KO ID','Type of Transaction'],axis = 1,inplace = True)
split
```

```python
# summing the amount of split transactions

df1 = split.groupby(by = ['KO ID / Type of Transaction','Transaction Date & Time','From Account','To Account'])['Amount'].sum().reset_index()
df1
```

```python
df1['Type of Transaction'] = [d.split(' / ')[1] for d in df1['KO ID / Type of Transaction']]
df1
```

```python
# merging the limits

merged = pd.merge(df1,limit,how = 'inner',on = 'Type of Transaction')
merged.drop('Type of Transaction',axis = 1,inplace = True)
merged
```

```python
# those split transactions whose total amount is less than or equal to the limit of transaction
# are suspicious transactions

suspicious_split_transactions = merged[merged['Amount']<=merged['Limit']]
suspicious_split_transactions
```

```python
# creating an excel file of the suspicious split transactions

suspicious_split_transactions.to_excel(r'Suspicious Split Transactions.xlsx',index = False)
```

#### Round Trip Transactions

```python
df
```

```python
from_account = df['From Account'].value_counts().reset_index()
from_account.rename(columns = {'index':'From Account','From Account':'Count'},inplace = True)
from_account
```

```python
to_account = df['To Account'].value_counts().reset_index()
to_account.rename(columns = {'index':'To Account','To Account':'Count'},inplace = True)
to_account
```

Clearly there are some round trip transactions happening.

```python
# we can identify the number of transactions happening by merging the two dataframes on the accounts

to_account.rename(columns = {'To Account':'From Account'},inplace = True)
to_account
```

```python
merged = pd.merge(from_account,to_account,how = 'inner',on = 'From Account')
merged
```

There are 591 such accounts which have round trip transactions going on.

```python
round_trip_transactions = pd.merge(df,merged,how = 'inner',on = 'From Account')
round_trip_transactions
```

There are 7198 such round trip transactions.

```python
# creating an excel file of the suspicious round trip transactions

round_trip_transactions.to_excel(r'Suspicious Round Trip Transactions.xlsx',index = False)
```
